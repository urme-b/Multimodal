```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
import warnings
from sklearn.cluster import KMeans
import numpy as np
from scipy.stats import f_oneway

# Suppress warnings
warnings.filterwarnings('ignore')

psychometric_01_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_03.csv')

# Function to process data for a participant
def process_data(participant_id, dataframes):
    for idx, df in enumerate(dataframes):
        # Convert datetime columns to datetime objects
        df['Question Start Time'] = pd.to_datetime(df['Question Start Time']).dt.tz_localize(None)
        df['Question Answer Time'] = pd.to_datetime(df['Question Answer Time']).dt.tz_localize(None)
        
        # Calculate the answer duration for each session
        df['answer_duration'] = (df['Question Answer Time'] - df['Question Start Time']).dt.total_seconds() / 60
        
        # Add columns to identify the participant and session
        df['Participant'] = f'P{participant_id:02d}'
        df['Session'] = f'Session {idx+1:02d}'
        
    return pd.concat(dataframes)

# Process data for each participant
participant_1_data = process_data(1, [psychometric_01_p1, psychometric_02_p1, psychometric_03_p1])
participant_2_data = process_data(2, [psychometric_01_p2, psychometric_02_p2, psychometric_03_p2])
participant_3_data = process_data(3, [psychometric_01_p3, psychometric_02_p3, psychometric_03_p3])
participant_4_data = process_data(4, [psychometric_01_p4, psychometric_02_p4, psychometric_03_p4])
participant_5_data = process_data(5, [psychometric_01_p5, psychometric_02_p5, psychometric_03_p5])
participant_6_data = process_data(6, [psychometric_01_p6, psychometric_02_p6, psychometric_03_p6])
participant_7_data = process_data(7, [psychometric_01_p7, psychometric_02_p7, psychometric_03_p7])
participant_8_data = process_data(8, [psychometric_01_p8, psychometric_02_p8, psychometric_03_p8])
participant_9_data = process_data(9, [psychometric_01_p9, psychometric_02_p9, psychometric_03_p9])
participant_10_data = process_data(10, [psychometric_01_p10, psychometric_02_p10, psychometric_03_p10])

# Concatenate all participants' data
all_data = pd.concat([participant_1_data, participant_2_data, participant_3_data,
                      participant_4_data, participant_5_data, participant_6_data,
                      participant_7_data, participant_8_data, participant_9_data,
                      participant_10_data])

# Trend Analysis: Mean answer durations across sessions
mean_durations = all_data.groupby(['Session', 'Participant'])['answer_duration'].mean().unstack()

# Plot mean answer durations across sessions for each participant
plt.figure(figsize=(15, 8))
sns.lineplot(data=mean_durations, markers=True, dashes=False)
plt.title('Mean Answer Duration Across Sessions for Each Participant')
plt.ylabel('Mean Answer Duration (minutes)')
plt.xlabel('Session')
plt.legend(title='Participant', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Clustering: Group participants based on their answer duration patterns
kmeans = KMeans(n_clusters=3, random_state=42).fit(mean_durations.T)
clusters = kmeans.labels_

# Add cluster information to the DataFrame
clustered_data = mean_durations.T.copy()
clustered_data['Cluster'] = clusters

# Plot clusters
plt.figure(figsize=(10, 6))
sns.heatmap(clustered_data.sort_values('Cluster').drop('Cluster', axis=1), annot=True, cmap='coolwarm')
plt.title('Participant Clusters Based on Answer Duration Patterns')
plt.show()

# Statistical Test: ANOVA to test if there's a significant difference across sessions
session_1 = all_data[all_data['Session'] == 'Session 01']['answer_duration']
session_2 = all_data[all_data['Session'] == 'Session 02']['answer_duration']
session_3 = all_data[all_data['Session'] == 'Session 03']['answer_duration']

anova_result = f_oneway(session_1, session_2, session_3)
print(f'ANOVA Result: F-value = {anova_result.statistic:.2f}, p-value = {anova_result.pvalue:.4f}')

# Interpret the ANOVA result
if anova_result.pvalue < 0.05:
    print("There is a statistically significant difference in answer durations across the sessions.")
else:
    print("There is no statistically significant difference in answer durations across the sessions.")

# Display Overall Trend Analysis
overall_trend = all_data.groupby('Session')['answer_duration'].agg(['mean', 'median']).reset_index()
print("Overall Trend Analysis")
print(tabulate(overall_trend, headers='keys', tablefmt='grid'))

# Plot the overall trend of mean and median answer durations across sessions
plt.figure(figsize=(10, 6))
sns.lineplot(data=overall_trend, x='Session', y='mean', marker='o', label='Mean')
sns.lineplot(data=overall_trend, x='Session', y='median', marker='s', label='Median')
plt.title('Overall Trend of Answer Durations Across Sessions')
plt.ylabel('Answer Duration (minutes)')
plt.xlabel('Session')
plt.legend(title='Statistic')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Individual Participant Trends
plt.figure(figsize=(15, 8))
sns.lineplot(data=all_data, x='Session', y='answer_duration', hue='Participant', marker='o')
plt.title('Answer Duration Trends Across Sessions for Each Participant')
plt.ylabel('Answer Duration (minutes)')
plt.xlabel('Session')
plt.legend(title='Participant', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Read the data
psychometric_01_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_03.csv')

# Function to process data for a participant
def process_data(participant_id, dataframes):
    for idx, df in enumerate(dataframes):
        # Convert datetime columns to datetime objects
        df['Question Start Time'] = pd.to_datetime(df['Question Start Time']).dt.tz_localize(None)
        df['Question Answer Time'] = pd.to_datetime(df['Question Answer Time']).dt.tz_localize(None)
        
        # Calculate the answer duration for each session
        df['answer_duration'] = (df['Question Answer Time'] - df['Question Start Time']).dt.total_seconds() / 60
        
        # Add columns to identify the participant and session
        df['Participant'] = f'P{participant_id:02d}'
        df['Session'] = f'Session {idx+1:02d}'
        
    return pd.concat(dataframes)

# Process data for each participant
participant_1_data = process_data(1, [psychometric_01_p1, psychometric_02_p1, psychometric_03_p1])
participant_2_data = process_data(2, [psychometric_01_p2, psychometric_02_p2, psychometric_03_p2])
participant_3_data = process_data(3, [psychometric_01_p3, psychometric_02_p3, psychometric_03_p3])
participant_4_data = process_data(4, [psychometric_01_p4, psychometric_02_p4, psychometric_03_p4])
participant_5_data = process_data(5, [psychometric_01_p5, psychometric_02_p5, psychometric_03_p5])
participant_6_data = process_data(6, [psychometric_01_p6, psychometric_02_p6, psychometric_03_p6])
participant_7_data = process_data(7, [psychometric_01_p7, psychometric_02_p7, psychometric_03_p7])
participant_8_data = process_data(8, [psychometric_01_p8, psychometric_02_p8, psychometric_03_p8])
participant_9_data = process_data(9, [psychometric_01_p9, psychometric_02_p9, psychometric_03_p9])
participant_10_data = process_data(10, [psychometric_01_p10, psychometric_02_p10, psychometric_03_p10])

# Concatenate all participants' data
all_data = pd.concat([participant_1_data, participant_2_data, participant_3_data,
                      participant_4_data, participant_5_data, participant_6_data,
                      participant_7_data, participant_8_data, participant_9_data,
                      participant_10_data])

# Aggregate data to get mean answer durations for clustering
mean_durations = all_data.groupby(['Participant', 'Session'])['answer_duration'].mean().unstack()

# Standardize the data before clustering
scaler = StandardScaler()
mean_durations_scaled = scaler.fit_transform(mean_durations)

# Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(mean_durations_scaled)

# Add cluster labels to the DataFrame
mean_durations['Cluster'] = clusters

# Calculate total timing, count, and standard deviation for each participant and session
session_stats = all_data.groupby(['Participant', 'Session'])['answer_duration'].agg(['count', 'sum', 'std']).reset_index()
session_stats.rename(columns={'count': 'Count', 'sum': 'Total Timing (minutes)', 'std': 'Standard Deviation (minutes)'}, inplace=True)

# Format the results to 2 decimal places but keep them as floats for plotting
session_stats['Total Timing (minutes)'] = session_stats['Total Timing (minutes)'].round(2)
session_stats['Standard Deviation (minutes)'] = session_stats['Standard Deviation (minutes)'].round(2)

# Display the results in table format
print("Total Timing During 3 Sessions for Each Participant")
print(tabulate(session_stats, headers='keys', tablefmt='grid'))

# Convert columns back to numeric for plotting
session_stats['Total Timing (minutes)'] = pd.to_numeric(session_stats['Total Timing (minutes)'])
session_stats['Standard Deviation (minutes)'] = pd.to_numeric(session_stats['Standard Deviation (minutes)'])

# Plot clusters
plt.figure(figsize=(10, 6))
sns.heatmap(mean_durations.drop('Cluster', axis=1).assign(Cluster=clusters).sort_values('Cluster').drop('Cluster', axis=1), annot=True, cmap='coolwarm', center=0)
plt.title('Participant Clusters Based on Answer Duration Patterns')
plt.show()

# Display the cluster assignments
print("Cluster Assignments:")
print(mean_durations[['Cluster']].reset_index())

# Visualize the clusters
plt.figure(figsize=(15, 8))
for cluster in range(kmeans.n_clusters):
    participants_in_cluster = mean_durations[mean_durations['Cluster'] == cluster].index
    cluster_data = mean_durations.loc[participants_in_cluster].drop('Cluster', axis=1).T
    for participant in cluster_data.columns:
        sns.lineplot(data=cluster_data[participant], label=f'Cluster {cluster+1}', marker='o')

plt.title('Mean Answer Duration Across Sessions by Clusters')
plt.ylabel('Mean Answer Duration (minutes)')
plt.xlabel('Session')
plt.legend(title='Cluster', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Visualize the results using a bar plot
plt.figure(figsize=(15, 8))
sns.barplot(x='Session', y='Total Timing (minutes)', hue='Participant', data=session_stats, errorbar=None)
plt.title('Total Timing for Each Session by Participant')
plt.ylabel('Total Timing (minutes)')
plt.xlabel('Session')
plt.legend(title='Participant', bbox_to_anchor=(1.05, 1), loc='upper left')
sns.despine(trim=True)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Plot each participant's 3 sessions time separately
plt.figure(figsize=(15, 8))
sns.catplot(x='Session', y='Total Timing (minutes)', col='Participant', data=session_stats, kind='bar', col_wrap=4, height=4, aspect=1)
plt.subplots_adjust(top=0.9)
plt.suptitle('Total Timing for Each Session by Participant')
plt.show()
```
