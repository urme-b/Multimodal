```python
import pandas as pd
import matplotlib.pyplot as plt

# Load all CSV files
psychometric_01_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p1 = pd.read_csv('/Users/urmebose/Desktop/Style/P01/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p2 = pd.read_csv('/Users/urmebose/Desktop/Style/P02/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p3 = pd.read_csv('/Users/urmebose/Desktop/Style/P03/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p4 = pd.read_csv('/Users/urmebose/Desktop/Style/P04/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p5 = pd.read_csv('/Users/urmebose/Desktop/Style/P05/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p6 = pd.read_csv('/Users/urmebose/Desktop/Style/P06/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p7 = pd.read_csv('/Users/urmebose/Desktop/Style/P07/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p8 = pd.read_csv('/Users/urmebose/Desktop/Style/P08/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p9 = pd.read_csv('/Users/urmebose/Desktop/Style/P09/Nuro/Psychometric_Test_Results_03.csv')

psychometric_01_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_01.csv')
psychometric_02_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_02.csv')
psychometric_03_p10 = pd.read_csv('/Users/urmebose/Desktop/Style/P10/Nuro/Psychometric_Test_Results_03.csv')

# Clean the data by removing the first column and converting 'Time(s)' to numeric
def clean_data(df):
    df = df.iloc[:, 1:]  # Remove the first column
    df['Time(s)'] = pd.to_numeric(df['Time(s)'], errors='coerce')  # Convert 'Time(s)' to numeric
    return df

psychometric_01_p1 = clean_data(psychometric_01_p1)
psychometric_02_p1 = clean_data(psychometric_02_p1)
psychometric_03_p1 = clean_data(psychometric_03_p1)

psychometric_01_p2 = clean_data(psychometric_01_p2)
psychometric_02_p2 = clean_data(psychometric_02_p2)
psychometric_03_p2 = clean_data(psychometric_03_p2)

psychometric_01_p3 = clean_data(psychometric_01_p3)
psychometric_02_p3 = clean_data(psychometric_02_p3)
psychometric_03_p3 = clean_data(psychometric_03_p3)

psychometric_01_p4 = clean_data(psychometric_01_p4)
psychometric_02_p4 = clean_data(psychometric_02_p4)
psychometric_03_p4 = clean_data(psychometric_03_p4)

psychometric_01_p5 = clean_data(psychometric_01_p5)
psychometric_02_p5 = clean_data(psychometric_02_p5)
psychometric_03_p5 = clean_data(psychometric_03_p5)

psychometric_01_p6 = clean_data(psychometric_01_p6)
psychometric_02_p6 = clean_data(psychometric_02_p6)
psychometric_03_p6 = clean_data(psychometric_03_p6)

psychometric_01_p7 = clean_data(psychometric_01_p7)
psychometric_02_p7 = clean_data(psychometric_02_p7)
psychometric_03_p7 = clean_data(psychometric_03_p7)

psychometric_01_p8 = clean_data(psychometric_01_p8)
psychometric_02_p8 = clean_data(psychometric_02_p8)
psychometric_03_p8 = clean_data(psychometric_03_p8)

psychometric_01_p9 = clean_data(psychometric_01_p9)
psychometric_02_p9 = clean_data(psychometric_02_p9)
psychometric_03_p9 = clean_data(psychometric_03_p9)

psychometric_01_p10 = clean_data(psychometric_01_p10)
psychometric_02_p10 = clean_data(psychometric_02_p10)
psychometric_03_p10 = clean_data(psychometric_03_p10)

# Combine data for each participant
combined_df_p1 = pd.concat([psychometric_01_p1, psychometric_02_p1, psychometric_03_p1], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p2 = pd.concat([psychometric_01_p2, psychometric_02_p2, psychometric_03_p2], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p3 = pd.concat([psychometric_01_p3, psychometric_02_p3, psychometric_03_p3], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p4 = pd.concat([psychometric_01_p4, psychometric_02_p4, psychometric_03_p4], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p5 = pd.concat([psychometric_01_p5, psychometric_02_p5, psychometric_03_p5], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p6 = pd.concat([psychometric_01_p6, psychometric_02_p6, psychometric_03_p6], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p7 = pd.concat([psychometric_01_p7, psychometric_02_p7, psychometric_03_p7], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p8 = pd.concat([psychometric_01_p8, psychometric_02_p8, psychometric_03_p8], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p9 = pd.concat([psychometric_01_p9, psychometric_02_p9, psychometric_03_p9], keys=['S01', 'S02', 'S03'], names=['Session'])
combined_df_p10 = pd.concat([psychometric_01_p10, psychometric_02_p10, psychometric_03_p10], keys=['S01', 'S02', 'S03'], names=['Session'])

# Combine all participants' data
combined_data = {
    'P01': combined_df_p1,
    'P02': combined_df_p2,
    'P03': combined_df_p3,
    'P04': combined_df_p4,
    'P05': combined_df_p5,
    'P06': combined_df_p6,
    'P07': combined_df_p7,
    'P08': combined_df_p8,
    'P09': combined_df_p9,
    'P10': combined_df_p10
}

# Calculate average and standard deviation for each participant
def calculate_stats(df):
    stats = df.groupby('Session')['Time(s)'].agg(['mean', 'std']).reset_index()
    stats.columns = ['Session', 'Average Duration (s)', 'STD of Duration (s)']
    return stats

stats = {p_str: calculate_stats(combined_data[p_str]) for p_str in combined_data}

# Plot the results for average duration
def plot_avg_duration(stats, title):
    fig, ax = plt.subplots()
    colors = plt.cm.get_cmap('tab10', len(stats))
    
    for i, p_str in enumerate(stats):
        x = stats[p_str]['Session']
        y_avg = stats[p_str]['Average Duration (s)']
        ax.plot(x, y_avg, marker='o', label=f'{p_str} - Average Duration (s)', color=colors(i))
    
    plt.title(title)
    plt.ylabel('Time (seconds)')
    plt.xlabel('Session')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot the results for standard deviation
def plot_std_duration(stats, title):
    fig, ax = plt.subplots()
    colors = plt.cm.get_cmap('tab10', len(stats))
    
    for i, p_str in enumerate(stats):
        x = stats[p_str]['Session']
        y_std = stats[p_str]['STD of Duration (s)']
        ax.plot(x, y_std, marker='o', label=f'{p_str} - STD of Duration (s)', color=colors(i))
    
    plt.title(title)
    plt.ylabel('Time (seconds)')
    plt.xlabel('Session')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_avg_duration(stats, 'Average Duration by Session (10 Participants)')
plot_std_duration(stats, 'STD of Duration by Session (10 Participants)')

# Display the results table for each participant
for p_str in stats:
    print(f"Statistics for {p_str}:\n", stats[p_str])
```
